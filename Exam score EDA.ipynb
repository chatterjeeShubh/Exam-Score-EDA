```python
# import python libraries

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt # visualizing data
%matplotlib inline
import seaborn as sns
import os
```

```python
df = pd.read_csv('D:\DA Docs\Exam data.csv')
df
```

<span style="font-size: 18px;font-weight: bold; text-decoration: underline">Cleaning Data </span>

```python
# Handle missing values
# Assuming missing values can be treated as a separate category, you can fill them with a label like 'Unknown'
df['WklyStudyHours'].fillna('Unknown', inplace=True)
df.describe()
```

```python
df.drop(columns=['Unnamed: '], inplace=True)
```

<span style="font-size: 18px;font-weight: bold; text-decoration: underline"> Plotting Univariate Data </span>

```python
# Define the categorical variables
categorical_vars = ['Gender', 'EthnicGroup', 'ParentEduc', 'LunchType']

# Create subplots for each categorical variable
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Create a bar plot for each categorical variable
for i, var in enumerate(categorical_vars):
    sns.countplot(x=var, data=df, ax=axes[i])
    axes[i].set_title(f'Bar Plot of {var}')
    axes[i].set_xlabel(var)
    axes[i].set_ylabel('Count')
    axes[i].tick_params(axis='x', rotation=45)

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

**--> In the above bar graphs, we can clearly see the count between both genders, Ethnic groups, Parental educational qualifications, and the type of Lunches students have.
Male and female are almost 50%-50%.
Highest no. of count is from ethnic group C & D.
Maximum of the parents are from some college, some high school or have a assosiate's degree.
Most students have standard lunch type.**

```python
# Create a figure and axis
fig, ax = plt.subplots(figsize=(8, 5))

# Distribution plot of MathScore
sns.kdeplot(df['MathScore'], shade=True, label='Math Score', ax=ax)

# Distribution plot of ReadingScore
sns.kdeplot(df['ReadingScore'], shade=True, label='Reading Score', ax=ax)

# Distribution plot of WritingScore
sns.kdeplot(df['WritingScore'], shade=True, label='Writing Score', ax=ax)

# Set axis labels and title
ax.set_xlabel('Score')
ax.set_ylabel('Density')
ax.set_title('Distribution of Scores')

# Add a legend
ax.legend()

# Show the plot
plt.show()
```

**--> In all the three types of scores, let it be Maths, Reading or Writing the majority of the scores are concentrated towards 60 to 80 marks range and its clear from the above shown lineplot.**

```python
# Set up subplots
fig, axes = plt.subplots(1, 3, figsize=(15, 3))

# Plot boxplots for each score with outliers
sns.boxplot(data=df['MathScore'], ax=axes[0], orient='h', showfliers=True)
axes[0].set_ylabel('Math Score')
axes[0].set_title('Math Score Distribution')

sns.boxplot(data=df['ReadingScore'], ax=axes[1], orient='h', showfliers=True)
axes[1].set_ylabel('Reading Score')
axes[1].set_title('Reading Score Distribution')

sns.boxplot(data=df['WritingScore'], ax=axes[2], orient='h', showfliers=True)
axes[2].set_ylabel('Writing Score')
axes[2].set_title('Writing Score Distribution')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

**--> In all the three above score boxplots, it clearly visible that the mean data is between 66 to 70.
The 80% of data lies between 58 to 80, there are some outliers that are between 0 and 20 marks.**

<span style="font-size: 18px;font-weight: bold;text-decoration: underline"> Plotting Bivariate Data </span>

```python
# Create a figure and three subplots
fig, axs = plt.subplots(1, 3, figsize=(12, 3))

# Box plot of MathScore by Gender
sns.boxplot(ax=axs[0], x='Gender', y='MathScore', data=df)
axs[0].set_title('Math Score')

# Box plot of ReadingScore by Gender
sns.boxplot(ax=axs[1], x='Gender', y='ReadingScore', data=df)
axs[1].set_title('Reading Score')

# Box plot of WritingScore by Gender
sns.boxplot(ax=axs[2], x='Gender', y='WritingScore', data=df)
axs[2].set_title('Writing Score')

# Adjust the spacing between subplots
plt.tight_layout()

# Display the merged plot
plt.show()
```

**--> The scores slightly vary wrt the genders. In Maths the mean score of females is 65 and for males it is 70, similarly in Reading the mean score of females is 73 and for males it is 65, in Writing the mean score of females is 75 and for males it is 63.**

```python
# Set up subplots
fig, axes = plt.subplots(1, 3, figsize=(17, 4))

# Plot boxplots for each score by ethnic group
sns.boxplot(data=df, x='EthnicGroup', y='MathScore', ax=axes[0])
axes[0].set_xlabel('Ethnic Group')
axes[0].set_ylabel('Math Score')
axes[0].set_title('Math Score Distribution by Ethnic Group')

sns.boxplot(data=df, x='EthnicGroup', y='ReadingScore', ax=axes[1])
axes[1].set_xlabel('Ethnic Group')
axes[1].set_ylabel('Reading Score')
axes[1].set_title('Reading Score Distribution by Ethnic Group')

sns.boxplot(data=df, x='EthnicGroup', y='WritingScore', ax=axes[2])
axes[2].set_xlabel('Ethnic Group')
axes[2].set_ylabel('Writing Score')
axes[2].set_title('Writing Score Distribution by Ethnic Group')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

**--> From the above boxplots we can conclude that Group E and Group D has slightly higher average scores than rest of the ethnic groups.**

```python
# Create a figure and axis
plt.figure(figsize=(12, 4))

# Boxplot of MathScore by ParentEduc
plt.subplot(1, 3, 1)
sns.boxplot(x='ParentEduc', y='MathScore', data=df, palette='Set2')
plt.xlabel('Parental Education')
plt.ylabel('Math Score')
plt.title('Math Score by Parental Education')
plt.xticks(rotation = 45)

# Boxplot of ReadingScore by ParentEduc
plt.subplot(1, 3, 2)
sns.boxplot(x='ParentEduc', y='ReadingScore', data=df, palette='Set2')
plt.xlabel('Parental Education')
plt.ylabel('Reading Score')
plt.title('Reading Score by Parental Education')
plt.xticks(rotation = 45)

# Boxplot of WritingScore by ParentEduc
plt.subplot(1, 3, 3)
sns.boxplot(x='ParentEduc', y='WritingScore', data=df, palette='Set2')
plt.xlabel('Parental Education')
plt.ylabel('Writing Score')
plt.title('Writing Score by Parental Education')
plt.xticks(rotation = 45)

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

**--> From the above boxplots its visible that parents with Master's degree and Bachelor's degree tend to have children with greater mscores in Math, reading as well as writing.**

<span style="font-size: 18px;font-weight: bold; text-decoration: underline"> Plotting Multivariate Data </span>

```python
# Set up subplots
fig, axes = plt.subplots(1, 3, figsize=(15,4))

# Plot violin plots for each score with ethnic group and gender bifurcation
sns.violinplot(data=df, x='EthnicGroup', y='MathScore', hue='Gender', ax=axes[0], split=True)
axes[0].set_ylabel('Math Score')
axes[0].set_title('Math Score Distribution by Ethnic Group and Gender')

sns.violinplot(data=df, x='EthnicGroup', y='ReadingScore', hue='Gender', ax=axes[1], split=True)
axes[1].set_ylabel('Reading Score')
axes[1].set_title('Reading Score Distribution by Ethnic Group and Gender')

sns.violinplot(data=df, x='EthnicGroup', y='WritingScore', hue='Gender', ax=axes[2], split=True)
axes[2].set_ylabel('Writing Score')
axes[2].set_title('Writing Score Distribution by Ethnic Group and Gender')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

**--> As we have seen in a boxplot above that Group E has a higher average score than the rest of the ethnic groups, now in this violinplot we can see that in Maths the male gender has higher contribution to for having more marks, in reading the females have higher contribution and in writing also female have higher contribution.Another point to notice is that in all males have higher average score in all the ethnic groups of Maths score whereas females have higher average in all the ethnic groups in both reading and writing score**

```python
# Select only numeric columns
numeric_df = df.select_dtypes(include=['int64', 'float64'])

# Calculate correlation matrix
correlation_matrix = numeric_df.corr()

# Print correlation matrix
print(correlation_matrix)

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix Heatmap')
plt.yticks(rotation =45)
plt.xticks(rotation =45)
plt.show()
```

**-->In this correlation matrix we get to know that all the scores have no correlation to the No. of siblings parameter.
There is also a strong correlation between Maths Score and Reading & Writing scores and vice versa**
